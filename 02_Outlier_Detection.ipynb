```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("df_analysis.csv")
df.head(10)
```

```python
df.describe()
```

```python
df= df.drop(['id','host_id','house_rules'],axis=1)
```

```python
df.columns.tolist()
```

```python
print(f"Number of rows detected: {len(df)}")
```

```python
print("OUTLIER DETECTION AND REMOVAL")
print("-"*30)
print("\nBefore Cleaning: ")

# FINDING PROBLEMATIC VALUES
print(f" -> minimum_nights < 0 : {(df['minimum_nights'] <0 ).sum()}")
print(f" -> minimum_nights > 365 : {(df['minimum_nights'] > 365).sum()}")
print(f" -> availability_365 < 0 : {(df['availability_365'] <0).sum()}")
print(f" -> availability_365 > 365 : {(df['availability_365'] > 365).sum()}")
print(f" -> price = 0 : {(df['price']==0).sum()}")
```

**1. REMOVE IMPOSSIBLE LOWER VALUES**

```python
df = df[df['minimum_nights'] >= 0]
df = df[df['availability_365'] >= 0]
df = df[df['price'] > 0]
df = df[df['service_fee'] >= 0]

print("\nRemoved negetive values")
print(f"Rows remaining : {len(df)}")
```

**2. REMOVE UNREALISTIC UPPER BOUNDS**

```python
df = df[df['minimum_nights'] <= 365]
print(f"Rows remaining: {len(df)}")
```

```python
df=df[df['reviews_per_month'] <= 30]
print(f"Rows removed: {len(df)}")
```

**3. REMOVE STATISTICAL OUTLIERS (price)**

```python
Q1_price = df['price'].quantile(0.25)
Q3_price = df['price'].quantile(0.75)
IQR_price = Q3_price - Q1_price

lower_bound = Q1_price - 1.5 * IQR_price
upper_bound = Q3_price + 1.5 * IQR_price

print(f"PRICE IQR ANALYSIS")
print(f" Q1 (25th percentile): ${Q1_price:.2f}")
print(f" Q3 (75th percentile): ${Q3_price:.2f}")
print(f" IQR: ${IQR_price:.2f}")
print(f" valid range: ${lower_bound:.2f} to ${upper_bound:.2f}")

# COUNT OUTLIERS BEFORE REMOVAL
price_outliers = ((df['price'] < lower_bound) | (df['price'] > upper_bound)).sum()
print(f" PRICE OUTLIERS COUNT: {price_outliers}")
```

```python
# remove price outliers
df= df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
print(f"Rows remaining: {len(df)}")
```

**4. REMOVE SERVICE FEE OUTLIERS(IQR METHOD)**

```python
Q1_fee = df['service_fee'].quantile(0.25)
Q3_fee = df['service_fee'].quantile(0.75)
IQR_fee = Q3_fee - Q1_fee

lower_bound = Q1_fee - 1.5 * IQR_fee
upper_bound = Q3_fee + 1.5 * IQR_fee

print(f"PRICE IQR ANALYSIS")
print(f" Q1 (25th percentile): ${Q1_fee:.2f}")
print(f" Q3 (75th percentile): ${Q3_fee:.2f}")
print(f" IQR: ${IQR_fee:.2f}")
print(f" valid range: ${lower_bound:.2f} to ${upper_bound:.2f}")

# COUNT OUTLIERS BEFORE REMOVAL
price_outliers = ((df['service_fee'] < lower_bound) | (df['service_fee'] > upper_bound)).sum()
print(f" PRICE OUTLIERS COUNT: {price_outliers}")

# remove price outliers
df= df[(df['service_fee'] >= lower_bound) & (df['service_fee'] <= upper_bound)]
print(f"Rows remaining: {len(df)}")
```

**5. DATA RANGE VALIDATION**

```python
df['last_review'] = pd. to_datetime(df['last_review'], errors='coerce')
today = pd.Timestamp.today()
airbnb_founded = pd.Timestamp('2008-01-01')
print(f"Valid data range: {airbnb_founded.date()} to {today.date()}")

# count date issues
future_dates = (df['last_review'] > today).sum()
old_dates= (df['last_review']< airbnb_founded).sum()
print('-'*50)
print(f"Future dates found: {future_dates}")
print(f"Dates before 2008: {old_dates}")
print("-"*50)
df=df[(df['last_review'] >= airbnb_founded) & (df['last_review'] <= today)]
print(f"Rows remaining: {len(df)}")

```

**5. FINAL SUMMARY**

```python
print("\nOUTLIER REMOVAL SUMMARY")
print("-"*30)

print('\nAFTER CLEANING: ')
print(f" -> Total rows: {len(df)}")
print(f" -> Rows removed: {86706 - len(df)}")
print(f" -> Percentage removed: {((86706 - len(df)) / 86706 * 100):.2f} %")

print("\nCLEANED DATA RANGES: ")
print(f" -> minimum_nights: {df['minimum_nights'].min():.0f} to {df['minimum_nights'].max():.0f}")
print(f" -> availability_365: {df['availability_365'].min():.0f} to {df['availability_365'].max():.0f}")
print(f" -> price: ${df['price'].min():.2f} to ${df['price'].max():.2f}")
print(f" -> reviews_per_month: {df['reviews_per_month'].min():.2f} to {df['reviews_per_month'].max():.2f}")
print(f" -> last_review: {df['last_review'].min()} to {df['last_review'].max()}")
```
